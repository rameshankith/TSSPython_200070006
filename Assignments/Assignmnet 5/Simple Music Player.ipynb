```python
from tkinter import *
from tkinter import filedialog            
from pygame import mixer                  

window=Tk()
window.geometry("400x250")
window.title("Simple Music Player")    #to run this player we need some sample mp3 files
mixer.init()                          
music_playing=False
paused=False

def add_one_song():
    song= filedialog.askopenfilename(filetypes=([('mp3 files' , '*.mp3')]))
    songs_list.insert(END,song)  #this will insert the name of file along with location of file
                                 #we can use replace to remove path and extensions if we know location of file
def add_more_songs():
    songs=filedialog.askopenfilenames(filetypes=([('mp3 files' , '*.mp3')]))
    for song in songs:
        songs_list.insert(END,song)
    
def remove_selected_song():
    song=songs_list.curselection()
    song=int(song[0])
    songs_list.delete(song)
    
    
def play_song():
    global music_playing
    global paused
    if not music_playing:
        song= songs_list.get(ACTIVE)
        mixer.music.load(song)
        mixer.music.play()
        music_playing=True
    if paused:
        mixer.music.unpause()
        music_playing=True

def stop_song():
    global music_playing
    mixer.music.stop()
    music_playing=False
    song=songs_list.curselection()
    song=int(song[0])
    songs_list.selection_clear(song)


def pause_song():
    global music_playing
    global paused
    mixer.music.pause()
    paused=True
    
def play_previous_song():
    current_song=songs_list.curselection()
    current_song_index=int(current_song[0])
    if (current_song_index>0):
        previous_song_index=current_song_index-1
        previous_song=songs_list.get(previous_song_index)
        mixer.music.load(previous_song)
        mixer.music.play()
        music_playing=True
        songs_list.selection_clear(current_song_index)
        songs_list.activate(previous_song_index)
        songs_list.selection_set(previous_song_index)
    
        
def play_next_song():
    current_song=songs_list.curselection()
    current_song_index=int(current_song[0])
    next_song_index=current_song_index+1
    next_song=songs_list.get(next_song_index)
    mixer.music.load(next_song)
    mixer.music.play()
    music_playing=True
    songs_list.selection_clear(current_song_index)
    songs_list.activate(next_song_index)
    songs_list.selection_set(next_song_index)
        
songs_list=Listbox(window, width=350,bg="white")
songs_list.pack(side=TOP,pady=10,padx=10)

menubar=Menu(window)
add_song=Menu(menubar,tearoff=0)
menubar.add_cascade(label ='Add songs to playlist', menu = add_song)
add_song.add_command(label="Add one song",command=add_one_song)
add_song.add_command(label="Add more songs",command=add_more_songs)

remove_song=Menu(menubar,tearoff=0)
menubar.add_cascade(label="Remove selected song from Playlist",menu=remove_song)
remove_song.add_command(label="Remove song",command=remove_selected_song)

btns_frame=Frame(window,width=350,height=50)
btns_frame.pack(side=BOTTOM)

previous=Button(btns_frame,text="Prev Song",bg="yellow",fg="black",command=play_previous_song)
previous.grid(row=0,column=0)

next1=Button(btns_frame,text="Next Song",bg="light blue",fg="black",command=play_next_song)
next1.grid(row=0,column=1,padx=2,pady=8)

play=Button(btns_frame,text="Play Song",bg="light green",fg="black",command=play_song)
play.grid(row=0,column=2,padx=2,pady=8)

pause=Button(btns_frame,text="Pause Song",bg="orange",fg="black",command=pause_song)
pause.grid(row=0,column=3,padx=2,pady=8)

stop=Button(btns_frame,text="Stop Song",bg="red",fg="black",command=stop_song)
stop.grid(row=0,column=4,padx=2,pady=8)

window.config(menu=menubar)
window.mainloop()
```
